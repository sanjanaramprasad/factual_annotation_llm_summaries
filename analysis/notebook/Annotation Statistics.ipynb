```python
'''
Average number of sentences 
Average length of sentences
Average number of inconsistent summaries 
'''
import spacy 
import numpy as np

nlp = spacy.load('en_core_web_sm')
def get_basic_stats(df):
    summaries = list(df['summary'].values)
    summ_sents = [list(nlp(summ).sents) for summ in summaries]
    source_sents = [list(nlp(art).sents) for art in list(df['article'].values)]
    print('Average number of sentences(Source)', np.mean([len(each) for each in source_sents]))
    print('Average number of sentences', np.mean([len(each) for each in summ_sents]))
    summary_label_pairs = {}
    for idx, row in df.iterrows():
        summary = row['summary']
        if summary not in summary_label_pairs:
            summary_label_pairs[summary] = []
        label_ann1 = row['inaccuracy_severity_ann1'] 
        label_ann1 = 1 if type(label_ann1) is str else 0
        label_ann2 = row['inaccuracy_severity_ann2']
        label_ann2 = 1 if type(label_ann2) is str else 0
        # avg_count = np.mean([label_ann1, label_ann2])
        if label_ann1 or label_ann1:
            summary_label_pairs[summary] += [row['nonfactual_sentence']]
    # print(summary_label_pairs)
    summary_label_pairs = {k : len(list(set(v))) for k ,v in summary_label_pairs.items()}
    print(len(summary_label_pairs))
    print('Average numer of inconsistent sentences', np.mean(list(summary_label_pairs.values())))
get_basic_stats(df_news)
```

```python
get_basic_stats(df_billsum)
```

```python
get_basic_stats(df_pubmed)
```
