```python
import json
import pandas as pd
```

```python
def read_file(filepath):
    df_dict = {}
    with open(filepath, 'r') as fp:
        data = json.load(fp)
    for instance in data:
        for k , v in instance.items():
            if k not in df_dict:
                df_dict[k] = []
            df_dict[k].append(v)
    return pd.DataFrame(df_dict)
```

```python
df = read_file('/home/ramprasad.sa/factual_annotation_llm_summaries/datasets/benchmark_llm/likert_evaluation_results.json')


```
